```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_steps = 291
n_gen = 300
R_list = np.linspace(1, 30, n_steps)
eta_0 = 900
alpha = 0.01

plt.rcParams['figure.figsize'] = [20,20]
eta = np.zeros((n_steps, n_gen))
eta[:,0] = eta_0
for i, R in enumerate(R_list):
    for t in range(n_gen-1):
        eta[i, t+1] = R*eta[i, t]*np.exp(-alpha*eta[i, t])
        
    plt.scatter(np.ones(100)*R, eta[i,-100:], c='k', s=1)

plt.show()
```

```python
R_rep2 = [40, 90, 120, 220]

plt.rcParams['figure.figsize'] = [10,5]
for R in R_rep2:
    plt.plot(eta[R,:41], label=f'R={R_list[R]}')
    
plt.legend()
plt.title('b) Representative population dynamics')
plt.xlabel('tau')
plt.ylabel('eta')
plt.show()
```
